## Import Data

```python
import kagglehub
import pandas as pd
import sqlite3
import os

path = kagglehub.dataset_download("rauffauzanrambe/fifa-world-cup-2026-player-performance-dataset")

project_dir = r"xxxxxxxxxxxxxxxxxxxx"
db_path = os.path.join(project_dir, "data_analyst.db")
conn = sqlite3.connect(db_path)

print("Mulai memasukkan tabel ke SQLite...")
for file in os.listdir(path):
    if file.endswith('.csv'):
        table_name = os.path.splitext(file)[0]

        csv_full_path = os.path.join(path, file)
        df_temp = pd.read_csv(csv_full_path)

        df_temp.to_sql(table_name, conn, if_exists='replace', index=False)
        print(f"-> Tabel '{table_name}' berhasil dibuat. Ukuran: {df_temp.shape}")

cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print("\nDaftar tabel resmi di database kamu:", cursor.fetchall())

conn.close()
```

```python
conn = sqlite3.connect(db_path)

df_wc = pd.read_sql_query("SELECT * FROM fifa_world_cup_2026_player_performance", conn)
```

## Data Cleaning

```python
df_wc.head(10)
```

```python
df_wc.describe()
```

```python
duplicate_count = df_wc.duplicated().sum()
print(f"Jumlah baris Duplikat : {duplicate_count}")

logical_duplicates = df_wc.duplicated(subset=['player_id', 'match_id']).sum()
print(f"Jumlah duplikasi logis (player_id + match_id): {logical_duplicates}")
```

```python
df_wc['match_date'] = pd.to_datetime(df_wc['match_date'], errors='coerce')

invalid_dates = df_wc['match_date'].isna().sum()
print(f"Jumlah tanggal tidak valid: {invalid_dates}")

print(f"Tanggal pertandingan pertama: {df_wc['match_date'].min()}")
print(f"Tanggal pertandingan terakhir: {df_wc['match_date'].max()}")
```

```python
df_wc['minutes_played'].describe()
```

```python
shots_anomaly = (df_wc['shots_on_target'] > df_wc['shots']).sum()
print(f"Jumlah baris error (Shots on Target > Total Shots): {shots_anomaly}")

passes_anomaly = (df_wc["successful_passes"] > df_wc["total_passes"]).sum()
print(f"Jumlah baris error (Successful passes > total passes): {passes_anomaly}")
```

```python
menit_aktual = df_wc.groupby('player_id')['minutes_played'].sum().reset_index(name='calculated_minutes')

menit_tercatat = df_wc.groupby('player_id')['total_minutes_tournament'].first().reset_index()

merge_check = menit_aktual.merge(menit_tercatat, on='player_id')
mismatch_minutes = merge_check[merge_check['calculated_minutes'] != merge_check['total_minutes_tournament']]

print(f"Jumlah pemain dengan total menit turnamen tidak sinkron: {len(mismatch_minutes)}")

mismatch_minutes['selisih'] = mismatch_minutes['calculated_minutes'] - mismatch_minutes['total_minutes_tournament']
print(mismatch_minutes.head())

match_counts = df_wc[df_wc['player_id'].isin(mismatch_minutes['player_id'])].groupby('player_id').size().reset_index(name='jumlah_match')
print(match_counts.head())

print("\nStatistik jumlah pertandingan per pemain secara keseluruhan:")
print(df_wc.groupby('player_id').size().describe())

print(df_wc[df_wc['player_id'] == 'P00001'][['player_name', 'team', 'opponent_team', 'tournament_stage', 'minutes_played', 'match_date']].head(10))
```

```python
df_wc['total_minutes_tournament'] = df_wc.groupby('player_id')['minutes_played'].transform('sum')
df_wc['total_goals_tournament'] = df_wc.groupby('player_id')['goals'].transform('sum')
df_wc['total_assists_tournament'] = df_wc.groupby('player_id')['assists'].transform('sum')

mismatch_check = df_wc[df_wc.groupby('player_id')['minutes_played'].transform('sum') != df_wc['total_minutes_tournament']]
print(f"Jumlah pemain tidak sinkron setelah diperbaiki: {len(mismatch_check)}")
```

```python
string_cols = df_wc.select_dtypes(include=['object']).columns
for col in string_cols:
    df_wc[col] = df_wc[col].astype(str).str.strip()

print("Daftar Posisi Pemain:")
print(df_wc['position'].unique())

print("\nDaftar Kaki Dominan:")
print(df_wc['preferred_foot'].unique())

print("\nDaftar Fase Turnamen:")
print(df_wc['tournament_stage'].unique())
```

```python
print("Statistik Akurasi Umpan:")
print(df_wc['pass_accuracy'].describe()[['min', 'max']])

print("\nStatistik Persentase Penyelamatan Kiper:")
print(df_wc['save_percentage'].describe()[['min', 'max']])
```

```python
bukan_kiper_save = df_wc[(df_wc['position'] != 'Goalkeeper') & (df_wc['save_percentage'] > 0)]
print(f"Jumlah pemain non-kiper yang punya skor save percentage: {len(bukan_kiper_save)}")

kiper_nol_save = df_wc[(df_wc['position'] == 'Goalkeeper') & (df_wc['save_percentage'] == 0.0)]
print(f"Jumlah match di mana kiper asli memiliki save percentage 0.0: {len(kiper_nol_save)}")
```

## Exploratory Data Analysis

```python
# 1. Statistik deskriptif untuk umur, tinggi, dan berat badan
print("Distribusi Fisik Pemain")
print(df_wc[['age', 'height_cm', 'weight_kg']].describe())

# 2. Sebaran persentase penggunaan kaki dominan
print("\nSebaran Kaki Dominan")
print(df_wc['preferred_foot'].value_counts(normalize=True) * 100)

# 3. Jumlah pemain unik berdasarkan posisi
# (Menggunakan drop_duplicates agar satu pemain hanya dihitung sekali)
print("\nJumlah Pemain Unik per Posisi")
print(df_wc.drop_duplicates(subset=['player_id'])['position'].value_counts())
```

```python
ofensif_posisi = df_wc.groupby('position')[['goals', 'assists', 'shots', 'expected_goals_xg']].mean()
print("Rata-rata Statistik Ofensif per Match berdasarkan Posisi")
print(ofensif_posisi)
```

```python
disiplin_posisi = df_wc.groupby('position')[['fouls_committed', 'yellow_cards', 'red_cards']].mean()
print("=== Rata-rata Statistik Disiplin per Match berdasarkan Posisi ===")
print(disiplin_posisi)
```

```python
df_pemain_unik = df_wc.drop_duplicates(subset=['player_id'])

top_scorers = df_pemain_unik[['player_name', 'team', 'position', 'total_goals_tournament']].sort_values(by='total_goals_tournament', ascending=False).head(5)

top_assisters = df_pemain_unik[['player_name', 'team', 'position', 'total_assists_tournament']].sort_values(by='total_assists_tournament', ascending=False).head(5)

print("TOP 5 SCORERS TOURNEY")
print(top_scorers.to_string(index=False))

print("\nTOP 5 ASSISTERS TOURNEY")
print(top_assisters.to_string(index=False))
```

```python
korelasi_rating = df_wc[['player_rating', 'goals', 'assists', 'shots_on_target', 
                         'pass_accuracy', 'defensive_actions', 'distance_covered_km']].corr()['player_rating']

print("Korelasi Metrik terhadap Player Rating")
print(korelasi_rating.sort_values(ascending=False))
```
